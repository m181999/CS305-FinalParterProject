```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
#Get data from Walgreens API
```

```python
postUrl = "https://services.walgreens.com/api/stores/search/v2"
header = {'Content-Type': 'application/json'}
payload= {"apiKey": "zfKgzqZlgYfE7wVEx4bCUoguBUc6ypvM","affId": "storesapi","lat": "41.881832","lng": "-87.623177","r": "10000","s": "10000", "requestType": "locator"}
jsonPayload = json.dumps(payload)
response = requests.post(url=postUrl,headers=header,data=jsonPayload)
print(response.text)
```

```python
#Load JSON into walgreensData dict
```

```python
walgreensData = json.loads(response.text)
```

```python
print(walgreensData)
```

```python
#set columns in walgreensDF (walgreens dataframe) input data from walgreensData gathered from API.
```

```python
new_columns = ['deliveryScheduling','deliveryFee','address','state','city','zip']
walgreensDF = pd.DataFrame(columns=new_columns)

for x in walgreensData['results']:
    walgreensDF.loc[len(walgreensDF)] = [x['deliveryScheduling'],x['deliveryFee'],x['store']['address']['street'],x['store']['address']['state'],x['store']['address']['city'],x['store']['address']['zip']]
```

```python
walgreensDF = walgreensDF.astype({'deliveryScheduling': int})
walgreensDF = walgreensDF.astype({'deliveryFee': float})
walgreensDF = walgreensDF.astype({'zip': int})
```
